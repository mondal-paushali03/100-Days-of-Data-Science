## **WEEK 4: DATA VISUALISATION USING PYTHON**

### **Day 1: Data Overview and Distribution Plots**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset('tips')

# Create subplots (2 plots in a row)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Task 1: Basic Overview (text output)
print(tips.info())
print(tips.head())

# Task 2: Histogram of total bill
sns.histplot(tips['total_bill'], kde=True, ax=axes[0])
axes[0].set_title('Total Bill Distribution')
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Frequency')

# Task 3: Boxplot of total bill by time
sns.boxplot(x='time', y='total_bill', data=tips, ax=axes[1])
axes[1].set_title('Total Bill by Time of Day (Lunch/Dinner)')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Total Bill')

plt.tight_layout()
plt.show()
```

### **Day 2: Exploring Total Bill and Tip**

```python
# Create subplots (2 plots in a row)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Task 1: Scatter plot of total bill vs tip
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', ax=axes[0])
axes[0].set_title('Total Bill vs Tip by Gender')
axes[0].set_xlabel('Total Bill')
axes[0].set_ylabel('Tip')

# Task 2: Violin plot of tip distribution by gender
sns.violinplot(x='sex', y='tip', data=tips, ax=axes[1])
axes[1].set_title('Tip Distribution by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Tip')

plt.tight_layout()
plt.show()
```

### **Day 3: Analyzing Tips by Day and Time**

```python
# Create subplots (2 plots in a row)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Task 1: Barplot of average tip by day
sns.barplot(x='day', y='tip', data=tips, ax=axes[0])
axes[0].set_title('Average Tip by Day')
axes[0].set_xlabel('Day')
axes[0].set_ylabel('Average Tip')

# Task 2: Boxplot of tip by day and time
sns.boxplot(x='day', y='tip', hue='time', data=tips, ax=axes[1])
axes[1].set_title('Tip Distribution by Day and Time')
axes[1].set_xlabel('Day')
axes[1].set_ylabel('Tip')

plt.tight_layout()
plt.show()
```

### **Day 4: Pairwise Relationships and Heatmap**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Task 1: Pairplot to explore relationships between numerical features
sns.pairplot(tips[['total_bill', 'tip', 'size']])
plt.suptitle('Pairwise Relationships: Total Bill, Tip, and Size', y=1.02)
plt.show()

# Task 2: Create subplots for the heatmap of the correlation matrix
fig, ax = plt.subplots(figsize=(8, 6))

# Select only numeric columns for the correlation matrix
numeric_columns = tips.select_dtypes(include=['number']).columns
corr = tips[numeric_columns].corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, ax=ax)
ax.set_title('Correlation Matrix of Tips Dataset')

plt.tight_layout()
plt.show()
```

### **Day 5: Exploring Tip Distribution by Gender and Time**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Task 1: Boxplot of tip distribution by gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='tip', data=tips)
plt.title('Tip Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Tip')
plt.show()

# Task 2: FacetGrid of total bill and tip by time
g = sns.FacetGrid(tips, col="time", height=6, aspect=1)
g.map(sns.scatterplot, 'total_bill', 'tip')
g.set_axis_labels('Total Bill', 'Tip')
g.set_titles("{col_name} Time")
g.tight_layout()
plt.show()
```

```python
# Create subplots (2 plots in a row)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Task 1: Barplot of tip by gender
sns.barplot(x='sex', y='tip', data=tips, ax=axes[0])
axes[0].set_title('Average Tip by Gender')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Average Tip')

# Task 2: Violin plot of tip distribution by size
sns.violinplot(x='size', y='tip', data=tips, ax=axes[1])
axes[1].set_title('Tip Distribution by Size')
axes[1].set_xlabel('Size')
axes[1].set_ylabel('Tip')

plt.tight_layout()
plt.show()
```
